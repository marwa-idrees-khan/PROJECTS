```python
import spacy
import pprint                  #print every word in new line
import statistics               
```

```python
spacysim=[]
simil=[]
wor3=[]
wor4=[]
```

```python
nlp=spacy.load('en_core_web_sm')          #loading spacy model ( english core web small )
```

```python
value=['farm','hotel','house','terrace','house','tram','river','stream','riverbank','pond','bakery','pharmacy','pub','restraurant','college','theatre','college','library','hell','heaven']
```

```python
value2=['cape','reef','volcano','hill','motorway','track','court','valley','park','groceryshop','university','school','barakahu','jilani','lahore','kernalStop','atrio','howday','lunar']
```

```python
valueSpacyDocs= [nlp(x) for x in value]
```

```python
conSpacyDocs =[nlp(x) for x in value2]
```

```python
print('hi', valueSpacyDocs)
```

```python
list=[]
for x in valueSpacyDocs:
    for y in conSpacyDocs:
        simi = x.similarity(y)
        spacysim.append((x,y,simi))

print("Similarity of target words with spacy")
pprint.pprint(spacysim)
```
