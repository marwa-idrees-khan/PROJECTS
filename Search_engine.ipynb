```python
# Import necessary libraries
import pandas as pd

# Create a dictionary with skincare product data
data = { 'Product Name': ['Hydrating Cleanser', 'Vitamin C Serum', 'Retinol Cream', 'Sunscreen SPF 50', 'Moisturizing Lotion', 
'Exfoliating Scrub', 'Anti-Aging Cream', 'Eye Cream', 'Face Mask', 'Toner',  'Night Cream', 'Acne Treatment', 'Brightening Serum', 'Facial Oil', 'Clay Mask', 
'Hydrating Mist', 'Lip Balm', 'Hand Cream', 'Body Lotion', 'Foot Cream','Micellar Water', 'Sheet Mask', 'BB Cream'],
'Brand': ['CeraVe', 'SkinCeuticals', 'Neutrogena', 'La Roche-Posay', 'Aveeno', 'St. Ives', 'Olay', 'Kiehl\'s', 'Origins', 'Thayers', 
'L\'Oreal', 'Clean & Clear', 'The Ordinary', 'Burt\'s Bees', 'Aztec Secret', 'Mario Badescu', 'ChapStick', 'Eucerin', 'Nivea', 'Gold Bond','Bioderma', 'TonyMoly', 'Garnier'],
'Price ($)': [12.99, 25.50, 30.00, 15.75, 18.00, 10.50, 40.00, 22.00, 14.00, 9.99, 35.00, 20.00, 28.00, 19.50, 12.00,  8.99, 3.50, 7.00, 11.00, 9.50,12.00,8.99, 3.50,],
'Key Ingredients': ['Hyaluronic Acid', 'Vitamin C', 'Retinol', 'Zinc Oxide', 'Shea Butter', 'Salicylic Acid', 'Peptides', 'Caffeine', 'Charcoal', 'Witch Hazel', 
'Ceramides', 'Benzoyl Peroxide', 'Niacinamide', 'Jojoba Oil', 'Kaolin Clay', 'Aloe Vera', 'Beeswax', 'Glycerin', 'Cocoa Butter', 'Urea','Micelles', 'Green Tea', 'SPF']}

# Convert the dictionary into a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to get the part of speech tag for lemmatization
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
# Function to preprocess text (tokenization, stemming, lemmatization)
def preprocess(text):
    tokens = word_tokenize(text)
    stemmed = [stemmer.stem(word) for word in tokens]
    lemmatized = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in stemmed]
    return lemmatized
```

```python
# Function to search for products
def search_products(query):
    query_tokens = preprocess(query)
    results = []
    for index, row in df.iterrows():
        product_tokens = preprocess(row['Product Name'])
        ingredient_tokens = preprocess(row['Key Ingredients'])
        if any(token in product_tokens for token in query_tokens) or any(token in ingredient_tokens for token in query_tokens):
            results.append(row)
    return pd.DataFrame(results)
    
```

```python
# Ask the user to enter a product name
d1= ("welcome")
print(d1)
query = input("Enter the product name to search: ")
results = search_products(query)

# Display the results

if not results.empty:
    print("Matching products found:")
    print(results)
else:
    print("No matching products found.")
```
